```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df_cnn = pd.read_csv('cnn_model_log.csv')
df_gru = pd.read_csv('gru_model_log.csv')
df_bilstm = pd.read_csv('bi_lstm_model_log.csv')
```

```python
def best_accuracy(df):
    return df.loc[df['val_accuracy'].idxmax()]
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df_cnn['epoch'], df_cnn['loss'], label='CNN Training Loss', linestyle='-')
plt.plot(df_gru['epoch'], df_gru['loss'], label='GRU Training Loss', linestyle='-')
plt.plot(df_bilstm['epoch'], df_bilstm['loss'], label='BiLSTM Training Loss', linestyle='-')
plt.title('Loss Over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df_cnn['epoch'], df_cnn['val_loss'], label='CNN Validation Loss', linestyle='-')
plt.plot(df_gru['epoch'], df_gru['val_loss'], label='GRU Validation Loss', linestyle='-')
plt.plot(df_bilstm['epoch'], df_bilstm['val_loss'], label='BiLSTM Validation Loss', linestyle='-')
plt.title('Loss Over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df_cnn['epoch'], df_cnn['accuracy'], label='CNN Training Accuracy', linestyle='-')
plt.plot(df_gru['epoch'], df_gru['accuracy'], label='GRU Training Accuracy', linestyle='-')
plt.plot(df_bilstm['epoch'], df_bilstm['accuracy'], label='BiLSTM Training Accuracy', linestyle='-')
plt.title('Accuracy Over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df_cnn['epoch'], df_cnn['val_accuracy'], label='CNN Validation Accuracy', linestyle='-')
plt.plot(df_gru['epoch'], df_gru['val_accuracy'], label='GRU Validation Accuracy', linestyle='-')
plt.plot(df_bilstm['epoch'], df_bilstm['val_accuracy'], label='BiLSTM Validation Accuracy', linestyle='-')
plt.title('Accuracy Over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid()
plt.show()
```

```python
# Get the best validation accuracies
best_cnn = best_accuracy(df_cnn)['val_accuracy']
best_gru = best_accuracy(df_gru)['val_accuracy']
best_bilstm = best_accuracy(df_bilstm)['val_accuracy']

# Data for the histogram
best_accuracies = [best_cnn, best_gru, best_bilstm]
models = ['CNN', 'GRU', 'BiLSTM']
```

```python
# Plot the histogram
plt.figure(figsize=(8, 6))
plt.bar(models, best_accuracies, color=['blue', 'green', 'orange'])
plt.title('Best Accuracy for each model')
plt.xlabel('Models')
plt.ylabel('Best Accuracy')
plt.ylim(0, 1)  # Assuming accuracy is between 0 and 1
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Annotate the bars
for i, acc in enumerate(best_accuracies):
    plt.text(i, acc + 0.01, f"{acc:.2f}", ha='center', va='bottom')

plt.show()
```
