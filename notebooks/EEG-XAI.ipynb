```python
import shap
import tensorflow as tf
from tensorflow.keras.models import load_model
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
def predict_fn(data):
    return model_lstm.predict(data)
```

```python
model_lstm = load_model('/kaggle/input/modellll/bi-lstm-model.h5')
```

```python
X = np.random.rand(100, 5)
```

```python
feature_names = ['Delta', 'Theta', 'Alpha', 'Beta', 'Gamma']
sample_data = X[:100]
sample_data_with_features = pd.DataFrame(sample_data, columns=feature_names)
```

```python
explainer = shap.KernelExplainer(predict_fn, sample_data_with_features)
```

```python
shap_values = explainer.shap_values(sample_data_with_features)[0]
```

```python
plt.figure(figsize=(12, 8))

shap.summary_plot(
    shap_values,
    sample_data_with_features,
    feature_names=feature_names,
    plot_type="dot",
    show=False
)

plt.title("SHAP Summary Plot for EEG Bands")
plt.tight_layout()
plt.show()
```
